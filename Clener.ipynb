```python
# --- Step 1: Import dependencies ---
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import joblib  # for saving the model

data = pd.read_csv("Data/Wheat/Wheat_Crop_Disease_Feature_Engineered_Encoded.csv")

X = data[['Stage_Code', 'Temp (Â°C)', 'Humidity (%)', 'Soil Moisture (%)',
          'Rainfall (mm)', 'Soil pH', 'Temp_Humidity_Interaction',
          'Rainfall_SoilMoisture_Ratio', 'Is_High_Humidity',
          'Is_High_Temp', 'Is_Acidic_Soil', 'Is_Alkaline_Soil']]

y = data['Disease_Code']

rf_model = RandomForestClassifier(
    n_estimators=300,
    random_state=42,
    class_weight='balanced'   
)
rf_model.fit(X, y)

# --- Step 5: Save the trained model for later use ---
joblib.dump(rf_model, "Models/wheat_disease_random_forest.pkl")

print("âœ… Model trained successfully and saved as 'wheat_disease_random_forest.pkl'")

# --- Optional: check training performance on the same dataset ---
train_pred = rf_model.predict(X)
from sklearn.metrics import accuracy_score, classification_report
acc = accuracy_score(y, train_pred)
print(f"Training Accuracy: {acc:.3f}")
print("\nClassification Report:\n", classification_report(y, train_pred, zero_division=0))
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
import joblib

# --- Step 2: Load Dataset ---
data = pd.read_csv("Data/Wheat/Wheat_Crop_Disease_Feature_Engineered_Encoded.csv")
```

```python
data['Crop Stage'].unique()
```

```python
data['Crop Disease'].unique()
```

```python
data['Crop Disease'].count()
```

```python

stage_disease_map = data.groupby('Crop Stage')['Crop Disease'].apply(list).reset_index()

# --- Step 4: Multi-label Encode Disease Labels ---
mlb = MultiLabelBinarizer()
y = mlb.fit_transform(stage_disease_map['Crop Disease'])

# --- Step 5: Aggregate Environmental Features per Stage ---
stage_env = data.groupby('Crop Stage')[[
    'Stage_Code', 'Temp (Â°C)', 'Humidity (%)', 'Soil Moisture (%)',
    'Rainfall (mm)', 'Soil pH', 'Temp_Humidity_Interaction',
    'Rainfall_SoilMoisture_Ratio', 'Is_High_Humidity',
    'Is_High_Temp', 'Is_Acidic_Soil', 'Is_Alkaline_Soil'
]].mean().reset_index(drop=True)

# --- Step 6: Train Random Forest Multi-Label Model ---
rf = RandomForestClassifier(n_estimators=300, random_state=42, class_weight='balanced')
multi_rf = MultiOutputClassifier(rf)
multi_rf.fit(stage_env, y)

# --- Step 7: Save Model and Label Binarizer ---
joblib.dump(multi_rf, "Models/wheat_multilabel_randomforest.pkl")
joblib.dump(mlb, "Models/wheat_disease_mlb.pkl")

print("âœ… Multi-label Random Forest model trained and saved successfully!")

# =============================================================
# ðŸŒ¾ Test Model with Sample Input
# =============================================================

# --- Step 8: Define Sample Input ---
sample_input = {
    "Stage_Code": 5,         # example: corresponds to a specific stage (replace as needed)
    "Temp (Â°C)": 26.5,
    "Humidity (%)": 75,
    "Soil Moisture (%)": 55,
    "Rainfall (mm)": 12,
    "Soil pH": 6.8
}

# --- Step 9: Compute Engineered Features for Input ---
sample_input["Temp_Humidity_Interaction"] = sample_input["Temp (Â°C)"] * sample_input["Humidity (%)"]
sample_input["Rainfall_SoilMoisture_Ratio"] = sample_input["Rainfall (mm)"] / (sample_input["Soil Moisture (%)"] + 1)
sample_input["Is_High_Humidity"] = int(sample_input["Humidity (%)"] > 80)
sample_input["Is_High_Temp"] = int(sample_input["Temp (Â°C)"] > 32)
sample_input["Is_Acidic_Soil"] = int(sample_input["Soil pH"] < 6.0)
sample_input["Is_Alkaline_Soil"] = int(sample_input["Soil pH"] > 7.5)

X_input = pd.DataFrame([sample_input])

# --- Step 10: Predict Probabilities for All Diseases (Fixed Version) ---
probas_list = []

for est in multi_rf.estimators_:
    p = est.predict_proba(X_input)
    
    # Handle single-class issues safely
    if isinstance(p[0], np.ndarray) and len(p[0]) == 2:
        prob = p[0][1]  # probability of disease presence
    else:
        prob = 0.0      # if only one class present in training
    probas_list.append(prob)

probas = np.array(probas_list)

# --- Step 11: Rank All Possible Diseases by Probability ---
disease_probs = pd.DataFrame({
    "Disease": mlb.classes_,
    "Probability": probas
}).sort_values(by="Probability", ascending=False)

# --- Step 12: Display Top Predictions ---
print("\nðŸŒ¾ Predicted Diseases for Given Stage & Conditions:")
print(disease_probs.head(10))  # Show top 10 diseases

# --- Step 13: Save Predicted Probabilities (Optional) ---
disease_probs.to_csv("Data/Wheat/Predicted_Disease_Rankings.csv", index=False)
print("\nðŸ“„ Predictions saved to 'Data/Wheat/Predicted_Disease_Rankings.csv'")
```
